```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
x1 = df[['sqft_living', 'bedrooms']]
y = df['price']
```

```python
x = sm.add_constant(x1)
model1 = sm.OLS(y, x).fit()
```

```python
model1.summary()
```

```python
yhat1 = 313 * x['sqft_living'] - 57070 * x['bedrooms'] + 79470
```

```python
while True:
    x_t1 = int(input('Living room sqft: '))
    x_t2 = int(input('Number of bedrooms: '))
    if x_t1 == 0:
        break
    else:
        y_pred1 = 313 * int(x_t1) - 57070 * int(x_t2) + 79470
        print('The house price is:', y_pred1)
        print('---------------------------------------------')
```

```python
x2 = df[['sqft_living', 'bedrooms', 'bathrooms', 'floors']]
y2 = df['price']
```

```python
x2_2 = sm.add_constant(x2)
model2 = sm.OLS(y, x2_2).fit()
```

```python
yhat2 = 309.32 * x2['sqft_living'] - 57850 * x2['bedrooms'] + 7853.52 * x2['bathrooms'] + 200.49 * x2['floors'] + 74670
```

```python
while True:
    x2_1 = int(input('Living room sqft: '))
    x2_2 = int(input('Number of bedrooms: '))
    x2_3 = int(input('Number of bathrooms: '))
    x2_4 = int(input('Floor: '))
    if x2_1 == 0:
        break
    else:
        y_pred2 = 309.32 * x2_1 - 57850 * x2_2 + 7853.52 * x2_3 + 200.49 * x2_4 + 74670
        print('The house price is:', y_pred2)
        print('------------------------------------------------------')
```
